```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\asus\Downloads\study_time_scores.csv")
data
```

```python
plt.scatter(data.Study_Hours,data.Exam_Score)
plt.show()
```

```python
import numpy as np



# Gradient Descent function
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].Study_Hours
        y = points.iloc[i].Exam_Score
        m_gradient += (-2/n) * x * (y - (m_now * x + b_now))
        b_gradient += (-2/n) * (y - (m_now * x + b_now))
    
    m = m_now - (L * m_gradient)
    b = b_now - (L * b_gradient)
    
    return m, b

m = 0
b = 0
L = 0.01  
epochs = 500

# Perform Gradient Descent
for i in range(epochs):
    m, b = gradient_descent(m, b, data, L)

# Plot results
plt.scatter(data.Study_Hours, data.Exam_Score, color="black", label="Data Points")
x_range = np.linspace(0, 10, 100)
plt.plot(x_range, m * x_range + b, color="red", label="Best Fit Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```
